```python
from surprise import Dataset, Reader, SVD, KNNBasic
from surprise.model_selection import cross_validate
```

```python
file_path = 'movie/ratings_small.csv'  # Replace with the correct path
reader = Reader(line_format='user item rating timestamp', sep=',', skip_lines=1)
data = Dataset.load_from_file(file_path, reader=reader)
```

```python
algos = {
    'PMF': SVD(biased=False),  # PMF is an unbiased version of SVD
    'User-based CF': KNNBasic(sim_options={'user_based': True}),
    'Item-based CF': KNNBasic(sim_options={'user_based': False})
}

# Perform 5-fold cross-validation and compute metrics
for name, algo in algos.items():
    cv_results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=True)
    avg_rmse = sum(cv_results['test_rmse']) / 5
    avg_mae = sum(cv_results['test_mae']) / 5
    print(f'Algorithm: {name}\nAverage RMSE: {avg_rmse}\nAverage MAE: {avg_mae}\n')
```

```python
import matplotlib.pyplot as plt
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import cross_validate

# Assuming you've loaded your dataset
# You might need to modify the reader according to your dataset
reader = Reader(line_format='user item rating timestamp', sep=',', skip_lines=1)
data = Dataset.load_from_file('movie/ratings_small.csv', reader=reader)

sim_options = {
    'cosine': {'name': 'cosine'},
    'MSD': {'name': 'msd'},
    'pearson': {'name': 'pearson'}
}

results = {}

for name, sim_option in sim_options.items():
    for user_based in (True, False):
        algo_type = 'User-based' if user_based else 'Item-based'
        algo = KNNBasic(sim_options={'name': sim_option['name'], 'user_based': user_based})
        cv_results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=False)
        key = f'{algo_type} CF ({name})'
        results[key] = (cv_results['test_rmse'].mean(), cv_results['test_mae'].mean())

# Plotting the results
rmse_values = [v[0] for v in results.values()]
mae_values = [v[1] for v in results.values()]
labels = list(results.keys())

x = range(len(labels))

plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
plt.bar(x, rmse_values, tick_label=labels)
plt.xticks(rotation=45)
plt.title('Average RMSE')
plt.ylabel('RMSE')

plt.subplot(1, 2, 2)
plt.bar(x, mae_values, tick_label=labels)
plt.xticks(rotation=45)
plt.title('Average MAE')
plt.ylabel('MAE')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import cross_validate

# Load your dataset
# Replace 'path_to_ratings.csv' with the actual path to your dataset
reader = Reader(line_format='user item rating timestamp', sep=',', skip_lines=1)
data = Dataset.load_from_file('movie/ratings_small.csv', reader=reader)

num_neighbors = [5, 10, 15, 20, 25, 30, 40, 50]
results = {}

for k in num_neighbors:
    for user_based in (True, False):
        algo_type = 'User-based' if user_based else 'Item-based'
        algo = KNNBasic(k=k, sim_options={'name': 'msd', 'user_based': user_based})
        cv_results = cross_validate(algo, data, measures=['RMSE', 'MAE'], cv=5, verbose=False)
        key = f'{algo_type} CF, k={k}'
        results[key] = (cv_results['test_rmse'].mean(), cv_results['test_mae'].mean())

# Plotting the results
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 5))

user_rmse = [results[key][0] for key in results if 'User-based' in key]
item_rmse = [results[key][0] for key in results if 'Item-based' in key]
user_mae = [results[key][1] for key in results if 'User-based' in key]
item_mae = [results[key][1] for key in results if 'Item-based' in key]

ax1.plot(num_neighbors, user_rmse, label='User-based CF', marker='o')
ax1.plot(num_neighbors, item_rmse, label='Item-based CF', marker='o')
ax1.set_title('RMSE vs Number of Neighbors')
ax1.set_xlabel('Number of Neighbors')
ax1.set_ylabel('Average RMSE')
ax1.legend()

ax2.plot(num_neighbors, user_mae, label='User-based CF', marker='o')
ax2.plot(num_neighbors, item_mae, label='Item-based CF', marker='o')
ax2.set_title('MAE vs Number of Neighbors')
ax2.set_xlabel('Number of Neighbors')
ax2.set_ylabel('Average MAE')
ax2.legend()

plt.show()
```

```python
from surprise import Dataset, Reader, KNNBasic
from surprise.model_selection import cross_validate

# Load your dataset
# Replace 'path_to_ratings.csv' with the actual path to your dataset
reader = Reader(line_format='user item rating timestamp', sep=',', skip_lines=1)
data = Dataset.load_from_file('movie/ratings_small.csv', reader=reader)

num_neighbors = [5, 10, 15, 20, 25, 30, 40, 50]
user_based_rmse = {}
item_based_rmse = {}

for k in num_neighbors:
    # User-based Collaborative Filtering
    user_algo = KNNBasic(k=k, sim_options={'name': 'msd', 'user_based': True})
    user_results = cross_validate(user_algo, data, measures=['RMSE'], cv=5, verbose=False)
    user_based_rmse[k] = user_results['test_rmse'].mean()

    # Item-based Collaborative Filtering
    item_algo = KNNBasic(k=k, sim_options={'name': 'msd', 'user_based': False})
    item_results = cross_validate(item_algo, data, measures=['RMSE'], cv=5, verbose=False)
    item_based_rmse[k] = item_results['test_rmse'].mean()

# Identifying the best K for each method
best_user_k = min(user_based_rmse, key=user_based_rmse.get)
best_item_k = min(item_based_rmse, key=item_based_rmse.get)

print(f'Best K for User-based CF: {best_user_k} with RMSE: {user_based_rmse[best_user_k]}')
print(f'Best K for Item-based CF: {best_item_k} with RMSE: {item_based_rmse[best_item_k]}')

# Comparing the best K values
is_same_k = best_user_k == best_item_k
print(f'Is the best K for User-based CF the same as for Item-based CF? {is_same_k}')
```
